```python
# データの読み込み
import pandas as pd

lifeExpect = pd.read_csv('/content/Life expectancy.csv')
lifeExpect.head(10)
```

```python
ct_conti = pd.read_csv('/content/List of Countries-Continents.csv')
ct_conti.head(10)
```

```python
# 2つのテーブルを内部結合
joined = pd.merge(lifeExpect, ct_conti, left_on='Entity',right_on='Country',how='left')
joined.head(10)
```

### 大陸（Continent）ごとの平均寿命の計算

```python
# code=nullで絞り込み
filtered = joined[joined['Code'].isnull()]
filtered.head(10)
```

```python
# 必要な列だけ絞り込み
selectCol = filtered[['Entity','Year','Period_life_expectancy']]
selectCol.head(10)
```

```python
#Yearを1950-2020で絞り込み
yearFilter = selectCol.query('Year >= 1950 & Year <= 2020')
yearFilter.head(10)
```

```python
# Year,Entityでグループ化して集計
agg = yearFilter.groupby(['Year','Entity']).mean('Period_life_expectancy')
agg.head(10)
```

```python
# 列名を変更
rename = agg.rename_axis(['Year', 'Continent']).rename(columns={'Period_life_expectancy':'Continent Avg'})
rename.head(3)
```

### 国（Country）ごとの平均寿命の計算

```python
# Year,Continent,Countryでグループ化して集計
agg_country = joined.groupby(['Year','Continent','Country']).mean('Period_life_expectancy')
agg_country.head(10)
```

```python
# 列名を変更
renameCountry = agg_country.rename(columns={'Period_life_expectancy':'Country Avg'})
renameCountry.head(10)
```

### 大陸と国を結合後、計算

```python
# 大陸ごとの集計と国ごとの集計を結合
ccAgg = pd.merge(renameCountry.reset_index(), rename, left_on=['Year','Continent'],right_on=['Year','Continent'],how='left')
ccAgg.head(10)
```

```python
# ccAggを1950~2020年に絞る
year_filter = ccAgg.query('Year >= 1950 & Year <= 2020')
year_filter.head(10)
```

```python
# 国の平均が大陸の平均を超えているかのフラグを作成
import numpy as np
year_filter['Flag'] = np.where(year_filter['Country Avg'] > year_filter['Continent Avg'], 1, 0)
year_filter.head(10)
```

```python
# 1950年のCountry Avgを追加
year_filter['1950'] = np.where(year_filter['Year']==1950,year_filter['Country Avg'],np.nan)
year_filter.head(10)
```

```python
# 2020年のCountry Avgを追加
year_filter['2020'] = np.where(year_filter['Year']==2020,year_filter['Country Avg'],np.nan)
year_filter.head(10)
```

```python
# Coutinent, Country, 1950, 2020をグループ化して集計
flgAgg = year_filter.groupby(['Continent','Country']).agg({'1950':'max','2020':'max','Flag':'sum','Continent':'count'})
flgAgg.head(10)
```

```python
# Continentの列名を変更
flgAggRename = flgAgg.rename(columns={'Continent':'cnt'})
flgAggRename.head(10)
```

```python
# 大陸の平均を上回った年の割合を計算
flgAggRename['% Years Above Continent Avg'] = round((flgAggRename['Flag'] / flgAggRename['cnt']) * 100,1)
flgAggRename
```

```python
# 変化率を計算
flgAggRename['% Change'] = round((flgAggRename['2020'] - flgAggRename['1950']) / flgAggRename['1950'] * 100,1)
flgAggRename
```

```python
# 変化率でランク付け
flgAggReset = flgAggRename.reset_index()
flgAggReset['Rank'] = flgAggReset.groupby('Continent')['% Change'].rank(ascending=False).round(0).astype(int)
flgAggReset
```

```python
# Rankを3位以内に絞り込み
rankFilter = flgAggReset.query('Rank <= 3')
rankFilter
```

```python
# 必要な列だけに絞る
output = rankFilter[['Continent','Rank','Country','% Years Above Continent Avg','% Change']]
output
```

```python
# 正解ファイルをインポート
solution = pd.read_csv('/content/2024W50.csv')
solution
```
